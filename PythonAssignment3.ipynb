```python
"""1.1 Write a Python Program to implement your own myreduce() function 
which works exactly like Python's built-in function reduce()
"""
def reduce(func,ist1):
    if(func=="mul"):
        prod=1
        for i in range(len(list1)):
            prod=prod*list1[i]
        return prod
    if(func=="add"):
        add=0
        for i in range(len(list1)):
            add=add+list1[i]
        return add
    
    
list1=[1,2,3,4,5]
output= reduce("mul",list1)
print(output)
output= reduce("add",list1)
print(output)
```

```python
"""
1.2 Write a Python program to implement your own myfilter() function 
which works exactly like Python's built-in function filter()
"""
def ffilter(func,list1):
    if(func=="odd"):
        temp=[]
        for i in range(0,len(list1)):
            if(list1[i]%2==1):
                temp.append(list1[i])
        return temp
    if(func=="even"):
        temp=[]
        for i in range(0,len(list1)):
            if(list1[i]%2==0):
                temp.append(list1[i])
        return temp
        
list1=[1,2,3,4,5,6]
output= ffilter("odd",list1)
print(output)
output= ffilter("even",list1)
print(output)
```
