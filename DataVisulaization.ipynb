```python
"""In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour  each data point with specific class.  
Hint:  
import numpy as np  
import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D  
from sklearn import decomposition  
from sklearn import datasets 
"""
#Iris dataset uploaded in github as csv

import numpy as np  
import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D  
from sklearn import decomposition  
from sklearn import datasets 
import pandas as pd
from mpl_toolkits import mplot3d
import math

df=pd.read_csv("C://Users//SITS//Videos//Ineuron//IRIS.csv")
df=df.head(12)
print(len(df))
print(df)
```

```python

print("*******************Contour Plot************")
fig = plt.figure()
ax = plt.axes(projection='3d')
X=df["sepal_length"]
Y=df["sepal_width"]
#print(X)
#print(Y)

Z = []
for i in range(144):
    Z.append(i)
#print(Z)

#print("**********Length*********")
#print(len(Z))

Z = np.array(Z).reshape(12,12)
#print(Z)

ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('3rd Dimension')
ax.set_title('3D contour')

plt.show()
```

```python

print("*******************Surface Plot************")
fig = plt.figure()
ax = plt.axes(projection='3d')
X=df["sepal_length"]
Y=df["sepal_width"]

#print(X)
#print(Y)

Z = []
for i in range(144):
    Z.append(i)
#print(Z)

#print("**********Length*********")
#print(len(Z))

Z = np.array(Z).reshape(12,12)
#print(Z)

ax.plot_surface(X, Y, Z,cmap='viridis', edgecolor='none')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('3rd Dimension')
ax.set_title('3D surface')

plt.show()
```

```python

print("*******************Scatter Plot************")
fig = plt.figure()
ax=plt.axes()
X=np.linspace(0,4,30)
Y=np.linspace(0,4,30)
X=df["sepal_length"]
Y=df["sepal_width"]
x=df["petal_width"]
y=df["petal_length"]

ax.scatter(X,Y,color='y')
ax.scatter(x,y,color='g')

ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_title('Scatter plot')

plt.show()


```
